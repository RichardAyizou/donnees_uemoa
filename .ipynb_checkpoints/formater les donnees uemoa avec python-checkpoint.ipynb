```python
import pandas as pd
import re
from io import StringIO
```

```python
with open('exportIndicateurs.xls', 'rb') as file:
    content = file.read()
```

```python
content_str = content.decode('utf-8')
contry_pattern = re.compile(r'\n\n(.*?)\s*;', re.DOTALL)
country = contry_pattern.findall(content_str)
```

```python
pattern = re.compile(rf'\n\n{re.escape(country[0])}(.*?)(?=\n\n|$)', re.DOTALL)
data_list = pattern.findall(content_str)
data_str = ''.join(data_list)
data_io = StringIO(data_str)
df1 = pd.read_csv(data_io, sep=';')
df1.reset_index(inplace=True)
columns1 = df1.iloc[0] 
df1 = df1[1:]
df1.columns = list(columns1)
df1 = df1.iloc[:, :-1]
df1 = df1.transpose()
columns1 = df1.iloc[1]
df1.columns = [col.replace('.', '').strip() for col in list(columns1)]
df1 = df1.reset_index().rename(columns={'index':'Annee'})
df1['Country'] = country[0]
merged = pd.DataFrame(columns=df1.columns)
merged
```

```python
# Isoler chaque pays
for c in country:
    pattern = re.compile(rf'\n\n{re.escape(c)}(.*?)(?=\n\n|$)', re.DOTALL)
    data_list = pattern.findall(content_str)
    data_str = ''.join(data_list)
    data_io = StringIO(data_str)
    df = pd.read_csv(data_io, sep=';')
    df.reset_index(inplace=True)
    columns = df.iloc[0] 
    df = df[1:]
    df.columns = list(columns)
    df = df.iloc[:, :-1]
    df = df.transpose()
    columns = df.iloc[1]
    df = df[2:]
    df.columns = [col.replace('.', '').strip() for col in list(columns)]
    df = df.reset_index().rename(columns={'index':'Annee'})
    df['Country'] = c
    merged = pd.concat([merged, df])

print('merged:', merged.shape)
merged.head()
```
